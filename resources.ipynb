### A few words on the local market for wildflowers

If regional biodiversity is to be preserved effectively, then the market for local, source-verified wildflowers must expand. In terms of demand, there is an increasing interest among private individuals in using local wildflowers to create natural gardens that provide habitat and food resources to the larger ecosystem. Public authorities are increasingly engaged in renaturalization projects that should use local, source-verified wildflowers. Indeed, there are an increasing number of companies offering to cut flowering meadows, dry the seed and offer larger scale wildtype wildflower mixes of variable genetic quality. The bias is towards grassland and wildflower meadow species for property developments and agricultural purposes, which leads to the fact that a consumer in Biel/Bienne would be hard pressed to find native wildflower seeds sourced locally.
