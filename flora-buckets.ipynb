## Risks and Opportunities

As the scientific and technical literature review previously laid out, we have already lost a significant amount of biodiversity and we risk losing much more in the years ahead: even common plant  species are becoming less widespread, and many rare species are dissapearing entirely on a local level. Preservation must occur on the local level, coordinated regionally, nationally and internationally, in order to prevent the local extinctions that impoverish the genetic diversity of remaining populations and are themselves the precursors to global extinctions. These efforts must be active measures that improve existing habitat, expand habitat and integrate the preservation of genetic diversity into its sourcing of plants.

These risks also present opportunities to review management practices by both public and private actors and reflect on what horticultural and gardening products can be brought tt market that can favour a reduction in carbon emmissions and an increase in ecological services.

## Hemiparasites
Hemiparasitic plants are a fascinating phenomenon and they are often becoming increasingly rare. They are however potentially valuable for urban restoration projects as they can reduce the vigour of aggressive members of Poaceae and Fabulaceae that can quickly dominate urban restoration projects.
Three hemiparasitic plants, Rhinanthus minor, Melampyrum arvense and M. nemorosum, will be grown in pots under the following configurations:
* Mature host plants Brizola media (poaceae) and Melilotus alba (fabaceae) are already present and the hemiparasite is planted alongside its roots
* A host plant (medicago lupulina and medicago sativa) and the hemiparasite are planted at the same time with the host plant germinated first
* No host plants
